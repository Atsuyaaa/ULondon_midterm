```python
# empty lists
address = []
room_layout = []
price = []
floor = []
year_built = []
space = []
sun_direction = []
balcony_space = []

# loop through all the links
for link in url:
    response = requests.get(link)

    soup = BeautifulSoup(response.content, 'html.parser')

    #address
    try:
        address.append(soup.find('span', class_='address').get_text(strip=True))
    except:
        address.append('')

    #room layout
    try:
        room_layout.append(soup.find('th', string='間取り').findNext('td').get_text(strip=True))
    except:
        room_layout.append('')

    #price
    try:
        price.append(soup.find('span', {'class' : 'price'}).get_text(strip=True))
    except:
        price.append('')

    #floor
    try:
        floor.append(soup.find('th', string='階数').findNext('td').get_text(strip=True))
    except:
        floor.append('')

    #year built
    try:
        year_built.append(soup.find('th', string='築年月').findNext('td').get_text(strip=True))
    except:
        year_built.append('')

    #space
    try:
        space.append(soup.find('th', string='専有面積').findNext('td').get_text(strip=True))
    except:
        space.append('')

    #sun direction
    try:
        sun_direction.append(soup.find('th', string='主要採光面').findNext('td').get_text(strip=True))
    except:
        sun_direction.append('')

    #balcony space
    try:
        balcony_space.append(soup.find('th', string='バルコニー面積').findNext('td').get_text(strip=True))
    except:
        balcony_space.append('')


    output = {"address" : address, 'room-layout' : room_layout, 'price' : price, 'floor' : floor, 'year built' : year_built, 
             'space' : space, 'sun direction' : sun_direction, 'balcony space' : balcony_space}
    


```

### Room layout Data : Add 1 more room when rooms have "S".
*"S" stands for a service room. Need to add 1 more room to justify*

```python
# Define a function to process the room-layout values
def process_layout(layout):
    # Extract the leading digit(s)
    num_rooms = int(layout[0])
    
    # If the layout contains 'S', add 1 to the number of rooms
    if 'S' in layout:
        num_rooms += 1
        
    return num_rooms

# Apply the function to the 'room-layout' column
df_combined['room-layout'] = df_combined['room-layout'].apply(process_layout)
```

###  Floor Data : Cut unnecessary values in balcony

```python
# Define a function to extract the floor number, including basement floors
def extract_floor(floor):
    if isinstance(floor, str):
        floor = floor.strip()  # Remove leading/trailing whitespace
        # Match regular floors and basement floors
        match = re.match(r'(B?\d+)階', floor)
        if match:
            return match.group(1)  # Return as string to keep all columns as objects
    return None  # Handle the case where the input is not a string or no match is found
# Apply the function to the 'floor(階)' column
df_combined['floor(階)'] = df_combined['floor(階)'].apply(extract_floor)

# Replace None values in the 'floor(階)' column with a specific value (e.g., 0)
df_combined['floor(階)'] = df_combined['floor(階)'].fillna(0)

# Convert the 'floor(階)' column to string type
df_combined['floor(階)'] = df_combined['floor(階)'].astype(str)
```

###  Year built Data : Cut unnecessary values in year built

```python
# Define a function to extract the year only
def extract_year(date_string):
    match = re.match(r'(\d{4})年', date_string)
    if match:
        year = match.group(1)
        return year
    else:
        return None  # or handle the case where no match is found

# Apply the function to the 'year built' column
df_combined['year built'] = df_combined['year built'].apply(extract_year)
```
